violinplot <br/>9: stripplot <br/>10: swarmplot <br/>11: catplot <br/>12: heatmap <br/> And some Graph Formatting (changing fig ure size, changing color, font , text size , rotating attribute names, adding plane and boxed text inside the graph).......

```python
import numpy as np
import pandas as pd
from numpy import array
from numpy.random import randn
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib 
import seaborn as sns
```

```python
# heatmap............
```

```python
df.corr() # finding co-relation in tabel form
```

```python
sns.heatmap(df.corr())
```

```python
sns.heatmap(df.corr(),annot=True,cmap='vlag',fmt='f') # using annote to  display complete value of that section....
```

```python
demo={'fontsize':12,'color':'k','alpha':0.6,'rotation':'vertical'
     ,'verticalalignment':'center','backgroundcolor':'w'}            # editing graph using dict()

sns.heatmap(df.corr(),annot=True,cmap='vlag',fmt='f',annot_kws=demo)
```

```python
demo1={'orientation':'horizontal','shrink':1,'extend':'min',
   'extendfrac':0.1               # creating a pencil bar
  }

sns.heatmap(df.corr(),cbar_kws=demo1)
```

```python
ax=sns.heatmap(df.corr(),linewidth=3,annot=True)
ax.tick_params(size=7,color='g',labelsize=10,labelcolor='b')   # changing some parameters of the heatmap
plt.title('heatmap of iris data',fontsize=20)
```

```python
R=sns.heatmap(df.corr())
R.set_xticklabels(R.get_xticklabels(), rotation=40) # rotating x axis attributes to 40
R.set_yticklabels(R.get_yticklabels(), rotation=40) # rotating y axis attributes to 40
```

```python
R=sns.heatmap(df.corr())
R.set_xticklabels(R.get_xticklabels(), rotation=40)
R.set_yticklabels(R.get_yticklabels(), rotation=40)

R.tick_params(labelcolor='r', labelsize='medium', width=3)
```

```python
plt.figure(figsize=(20,10))
txt=sns.barplot(x='species',y='petal_width',data=df)
txt.text(0.001,1, 'demo text here', fontsize=15)#    Inserting a normal text in the graph



txt.text(0.5, 2, 'Demo text in box add desc', style='italic',fontsize=20, # adding a boxed text in the graph....
        bbox={'facecolor': 'blue', 'alpha': 0.5, 'pad': 30})
```
